```python
import areal_interpolation as areal
import geopandas as geopd
import pandas as pd
```

```python
census_data = pd.read_csv("/home/data/census/nyc/acs/acs_wide.csv")
census_geo = geopd.read_file("/home/data/census/nyc/geo/nyc_tracts.geojson")
school_dis = geopd.read_file("https://data.cityofnewyork.us/api/geospatial/r8nu-ymqj?method=export&format=GeoJSON")
police_pre = geopd.read_file("https://data.cityofnewyork.us/api/geospatial/78dh-3ptz?method=export&format=GeoJSON")
```

```python
schools_interpolated = areal.main(census_geo, census_data, school_dis, "school_dist")
schools_interpolated
```

```python
police_interpolated = areal.main(census_geo, census_data, police_pre, "precinct")
police_interpolated
```

```python
police_pre["id"] = "police_" + police_pre["precinct"]
school_dis["id"] = "school_" + school_dis["school_dist"]
```

```python
combined_polygons = police_pre[["id", "geometry"]].append(school_dis[["id","geometry"]])
combined_polygons = combined_polygons.to_crs("EPSG:2263")
```

```python
census_geo = census_geo.to_crs("EPSG:2263")
census_geo["census_area"] = census_geo.area
intersected_polygons = combined_polygons.overlay(census_geo, how = "intersection")
intersected_polygons["intersected_area"] = intersected_polygons.area
intersected_polygons["areal_weight"] = intersected_polygons["intersected_area"] / intersected_polygons["census_area"]
```

```python
areal_weights = intersected_polygons[["id", "GEOID", "areal_weight"]]
```

```python
intersected_polygons[intersected_polygons["GEOID"] == "36061004700"]
```

```python
census_geo[census_geo["GEOID"] == "36061004700"].area
```

```python
census_data["GEOID"] = census_data["GEOID"].astype(str)
census_areal = census_data.merge(areal_weights, how = "inner")
census_values = census_areal.drop(["GEOID", "NAME", "areal_weight"], axis = 1)
census_values.iloc[:, census_values.columns != "id"] = census_values.iloc[:, census_values.columns != "id"].apply(pd.to_numeric, errors='coerce')
sums = pd.concat([census_values.iloc[:, census_values.columns != "id"].multiply(census_areal["areal_weight"], axis = "index"),census_values["id"]], axis = 1).groupby("id").sum()
```
